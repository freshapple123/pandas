```python
scientists = pd.DataFrame({
        "Name": ["Rosaline Franklin", "William Gosset"],
        "Occupation": ["Chemist", "Statistician"],
        "Born": ["1920-07-25", "1876-06-13"],
        "Died": ["1958-04-16", "1937-10-16"],
        "Age": [37, 61],
})
```

```python
scientists
```

```python
sci = pd.read_csv('../data/scientists.csv')
```

```python
sci
```

```python
sci.info
```

```python
sci.describe()
```

```python
sci['Died_dt'] = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
```

```python
sci['Born_dt'] = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
```

```python
sci.info()
```

```python
sci['Age'].sample(frac=1,random_state=42)
```

```python
sci
```

```python
sci['age_days'] = sci['Died_dt'] - sci['Born_dt']
```

```python
sci
```

```python
sci['age_days'].dt
```

```python
sci['age_days'].iloc[0]
```

```python
sci['age_days'].dt.days
```

```python
sci = sci.assign(
    age_days_assign=sci['Died_dt'] - sci['Born_dt'],
    age_year_assign=(sci['age_days'].dt.days / 365).apply(np.floor)
)
```
